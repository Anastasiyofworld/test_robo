Сразу стоит отметить, что данные загрузилились в строковом формате. А названия колонок - в числовом. Исправим это.

Так как на каждом листе с информацией у нас одинаковые названия колонок, мы сразу можем написать функцию, которую будем применять для каждого листа. 

Мы хотим, чтобы функция:
1. Заменила название колонки `Country name` (с пробелом и без) на `country`
2. Добавляла к названию остальных колонок (которые названы годом, по которому предоставлена информация) дополнительное описание в формате `2005` --> `population_2005`
3. Удаляла колонке с названиями страны дополнительные пробелы в конце строки и заменяла точку и запятую в названии на пробел. 
    - Если этого не сделать, при объединении таблиц мы будем получать неявные дубликаты, так как программа распознает названия стран без пробела и с пробелом за разные страны.

```python
def country_without_commas(data):
    '''удаляет лишние символы в колонке country'''
    data['country'] = data['country'].apply(lambda x: x.replace(',', ' ')).apply(lambda x: x.replace('.', ' ')).apply(str.strip)

def rename_columns(data, add_title):
    '''Функция:
    1) заменяет в таблице data
    название колонки Country name на сountry
    2) добавляет название add_title к названию колонки:
    2005 --> population_2005
    3) удаляет лишние символы в колонке country'''
    
    data.rename(columns={'Country name': 'country',
                        'Country name ': 'country'}, inplace=True)
    
    data.columns = [add_title + str(i).strip() if i != 'country' else 'country' for i in data.columns]
    
#     data['country'] = data['country'].apply(lambda x: x.replace(',', ' ')).apply(lambda x: x.replace('.', ' ')).apply(str.strip)

    country_without_commas(data)
```

```python
rename_columns(population, 'population_')
```

Исправим данные, которые загрузились в текстовом формате вместо числового. 

Напишем функцию, которая будет принимать значения в формате `6,275,076,016.5` и возвращать в `6275076016.5` - число с плавающей точкой.

```python
def to_float(data):
    for column in data.columns:
        if column != 'country':
            data[column] = data[column].str.replace(',', '').apply(float)
```

```python
to_float(population)
```

Напишем функцию, которая работает с библиотекой `pycountry_convert` и превращает названия стран в названия континентов.

```python
def country_to_continent(country_name):
    try:
        country_alpha2 = pc.country_name_to_country_alpha2(country_name)
        country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
        country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
        return country_continent_name
    except Exception:
        return None
```

```python
df['country'] = df['country'].str.replace('St ', 'Saint ').str.replace('  ', ' ').str.replace('  ', ' ')
```

```python
df['continent'] = df['country'].apply(country_to_continent)
```

У них сложные названия. Придется дополнить функцию для распознавания континента. Теперь она возьмет все слова по отдельности, и если они не являются артиклями и частицами, попытается распознать континент.

```python
def by_word(country_name):
    for word in country_name.split():
        if len(word) > 3:
            name = country_to_continent(word)
            if name is not None:
                return name
    return None
```

Применим данную функцию только к тем странам, у которых в колонке с континентом пропуск.

Для этого создадим функцию, которая будет брать нужные нам колонки и считать относительное изменение.

Например: колонка с именем `population_2005_2006` будет характеризовать, на какую долю изменилась численность населения в 2006 году относительно 2005.

```python
def relative_change(year1, year2, name):
    
    list_of_name = [
        str(name) + '_' + str(year1),
        str(name) + '_' + str(year2),
        str(name) + '_' + str(year1) + '_' + str(year2)
    ]
    
    df[list_of_name[2]] = (df[list_of_name[1]] - df[list_of_name[0]]) / df[list_of_name[0]]
```

```python
for name in [
    'population',
    'electric_power',
    'energy_imports',
    'birth_rate',
    'internet_users',
    'co2',
    'tax_revenue',
    'gdp',
    'agriculture',
    'gross_savings'
]:
    for year in range(2005, 2009):
        relative_change(year, year+1, name)
```

```python
def show_electric(countries, legend_loc1=0, legend_loc2=0, figsize=(13, 4), subplots=[1, 2]):
    fig, axes = plt.subplots(subplots[0], subplots[1], figsize=figsize)
    short.loc[short.index.isin(countries)][ELECTRIC_LIST].plot(kind='bar', 
                                                           ax=axes[0],
                                                           logy=True,
                                                           rot=0)
    axes[0].legend(LEGEND_YEAR, loc=legend_loc1)

    axes[0].set_ylabel('Потребление электроэнергии \n(кВтч на душу населения)')
    short.loc[short.index.isin(countries)][ELECTRIC_CHANGE_LIST].plot(kind='bar',
                                                                  ax=axes[1], 
                                                                  rot=0)
    axes[1].legend(LEGEND_YEAR_YEAR, loc=legend_loc2)
    axes[1].set_ylabel('Относительное изменение \nпотребления электроэнергии, доли')
    plt.suptitle(f'Изменение потребления электроэнергии')
    plt.tight_layout()
    plt.show()
```

```python
def show_electric_1(countries, legend_loc1=0, legend_loc2=0, figsize=(13, 4)):
    subplots=[1, len(countries)]
    
    fig, axes = plt.subplots(subplots[0], subplots[1], figsize=figsize)
    count = 0
    for i in countries:

        short.loc[short.index.isin([i])][ELECTRIC_LIST].plot(kind='bar', 
                                                               ax=axes[count] if len(countries) > 1 else axes ,
                                                               logy=True,
                                                               rot=0)
        axes[count].legend(LEGEND_YEAR, loc=legend_loc1)
        count += 1

    axes[0].set_ylabel('Потребление электроэнергии \n(кВтч на душу населения)')
#     axes[0].legend(LEGEND_YEAR, loc=legend_loc1)
    

    plt.suptitle(f'Изменение потребления электроэнергии')

    plt.tight_layout()    
    plt.show()
```

```python
def show_electric_2(countries, legend_loc1=0, legend_loc2=0, figsize=(13, 4)):
    fig, axes = plt.subplots(1, 1, figsize=figsize)
    short.loc[short.index.isin(countries)][ELECTRIC_CHANGE_LIST].plot(kind='bar',
                                                                        ax=axes,
                                                                        rot=0)
    axes.legend(LEGEND_YEAR_YEAR, loc=legend_loc2)
    axes.set_ylabel('Изменение относительного \nпотребления электроэнергии, доли')
    plt.suptitle(f'Относительное изменение потребления электроэнергии')
    plt.tight_layout()
    plt.show()
```

### Анализ влияния кризиса на страну на примере США

Как мы знаем, Соединенные Штаты - страна с самой развитой экономикой, с нее начался кризис. И было бы интересно сделать комплексный анализ и посмотреть, как кризис отразился на ее разных сферах.

Для этого возьмем данные по Соединенным Штатам и визуализируем их.

```python
USA = short[short.index =='United States'].iloc[:, :50]
```

```python
def usa_show(parameter, ax):
    USA.loc[ :, USA.columns.str.startswith(parameter)].plot(
        kind='bar',
        rot=0, 
        ax=axe[ax],
    )
    axe[ax].legend(False)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 8))
axe = axes.ravel()

USA.loc[ :, USA.columns.str.startswith('agriculture')].plot(
        kind='bar',
        rot=0, 
        ax=axe[0],
    )
axe[0].set_ylabel('Сельское хозяйство, \nдобавленная стоимость (%ВВП)')
axe[0].set_title('Сельское хозяйство')

USA.loc[ :, USA.columns.str.startswith('co2')].plot(
        kind='bar',
        rot=0, 
        ax=axe[1],
    )
axe[1].set_ylabel('Выбросы $CO_2$, \nметрические тонны на душу населения')
axe[1].set_title('Выбросы $CO_2$')

USA.loc[ :, USA.columns.str.startswith('electric_power')].plot(
        kind='bar',
        rot=0, 
        ax=axe[2],
    )
axe[2].set_ylabel('Потребление электроэнергии, \n(кВтч на душу населения)')
axe[2].set_title('Потребление электроэнергии')

USA.loc[ :, USA.columns.str.startswith('energy_imports')].plot(
        kind='bar',
        rot=0, 
        ax=axe[3],
    )
axe[3].set_ylabel('Импорт энергии, нетто \n(% от потребления энегргии)')
axe[3].set_title('Импорт энергии')

USA.loc[ :, USA.columns.str.startswith('tax_revenue')].plot(
        kind='bar',
        rot=0, 
        ax=axe[4],
    )
axe[4].set_ylabel('Налоговые поступления\n(% от ВВП)')
axe[4].set_title('Налоговые поступления')

USA.loc[ :, USA.columns.str.startswith('internet_users')].plot(
        kind='bar',
        rot=0, 
        ax=axe[5],
    )
axe[5].set_ylabel('Количество интернет-пользователей\n(на 100 чел)')
axe[5].set_title('Количество интернет-пользователей')
axe[5].legend(list(range(2005, 2010)), loc=4)
for i in range(6):
    axe[i].xaxis.set_visible(False)
for i in range(5):
    axe[i].get_legend().remove()

plt.suptitle('Соединенные Штаты в 2005-2009 гг')
plt.show()
```

#### **Вывод**

Итак, на данных графиках мы видим ожидаемую корреляцию между сельским хозяйством, потреблением энергии и выбросами $CO_2$.

Сельское хозяйсто, будучи одним из наиболее загрязняющих производств в плане выбросов $CO_2$ отреагировало на кризис снижением добавленной стоимости, что привело к снижению выбросов $CO_2$. Снижение потребления электроэнергии на фоне кризиса также положительно сказалось на экологической обстановке в США.

На второй линии графиков видна некоторая зависимость между импортом энергии и налоговыми поступлениями. 

На фоне снижения импорта энергии мы видим снижение $CO_2$, так как в США налоги взымаются не только с самих производств, но и на выбрасываемый ими $CO_2$, мы видим, что при небольшом снижении используемой импортируемой энергии, аграрного сектора, и вследстви этого $CO_2$ намного более выраженное снижение поступаемых налогов в 2009 году.

Можно предположить, что потенциальное снижение интернет-пользователей в 2009 году так же снизило количество потребляемой энергии, так как в 2009 году огромное количество работников банковского сектора, основная работа которых была связана с работой через Интернет, потеряла работу, что привело к логичному снижению интернет-пользователей, и как следствие снижению потребления электроэнергии.
